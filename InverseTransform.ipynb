```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from collections import defaultdict

# initialize list of lists
data = [["tom", 10,'Pakistan'], ['nick', 15,'Canada'], ['juli', 14,'USA'],['Imran',14,'Australia']]
  
# Create the pandas DataFrame
data = pd.DataFrame(data, columns = ['Name', 'Age','Country'])
print(data )
# print dataframe.
data.dtypes
```

```python
string_columns = list(data.select_dtypes(include=['object']).columns)
```

```python
string_columns= data[string_columns]
```

```python
d = defaultdict(LabelEncoder)
fit = string_columns.apply(lambda x: d[x.name].fit_transform(x))
```

```python
fit
```

```python
inversed_df=fit.apply(lambda x: d[x.name].inverse_transform(x))
inversed_df
```

```python
pd.concat([data["Age"], inversed_df], axis=1)
```
